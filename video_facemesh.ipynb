```python
# pip install opencv-python
# pip install mediapipe
import cv2
import os
import mediapipe as mp
from PIL import Image
import time

mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh()

cap = cv2.VideoCapture("gc4.mov")
# image=cv2.imread(".jpg")
final_arr=[]
final_all_array=[]
count=0
c=0
currentframe = 0
# used to record the time when we processed last frame
prev_frame_time = 0
# used to record the time at which we processed current frame
new_frame_time = 0

while True:
    ret,image = cap.read()
    #condition for stoppinf after all frames of video
    if ret is not True:
        break
    h,w,l= image.shape
    print("Height and width", h, w)
    
    #coverting image from bgr to rgb
    rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    result = face_mesh.process(rgb_image)
    print(result)
    try:
        # creating a folder named data
        if not os.path.exists('data'):
            os.makedirs('data')
        # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')
    
    #showing all landmarks
    for facial_landmarks in result.multi_face_landmarks:
        for i in range(0,468):
            pt1 = facial_landmarks.landmark[i]
            x= int(pt1.x * w)
            y= int(pt1.y * h)
            z= int(pt1.z * l)
            
            final_arr.append([x,y])
                        
#             print("x, y, z", x,y,z )
            lip_set={#outer lip line
                     61, 185, 40, 39, 37, 0, 267, 269, 270, 409, 291, 375, 321, 405, 314, 17, 84, 181, 91, 146,
                     #inner lip line
                     78, 191, 80,81, 82, 13, 312, 311, 310, 415, 308, 324, 318, 402, 317, 14, 87, 178, 88, 95}
            c_set={207,427,431,211}
            if i in lip_set:
                cv2.circle(image,(x,y), 2, (100,100,0),-1)
            if i==467:
                p1_x=final_arr[206][0]
                p1_y=final_arr[206][1]
                p2_x=final_arr[426][0]
                p2_y=final_arr[426][1]
                p3_x=final_arr[431][0]
                p3_y=final_arr[431][1]
                p4_x=final_arr[211][0]
                p4_y=final_arr[211][1]
                lip_crop = image[p1_y:p4_y,p1_x:p2_x]   # Top left part of the photo
                cv2.imshow("lip_crop",lip_crop)     # It will show cropu1 part in a window 

        name = './data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)
        cv2.imwrite(name, lip_crop)
        currentframe += 1
    final_all_array.append(final_arr)
#     print(final_all_array)
    cv2.imshow("Image", image)
    final_arr.clear()
     # resizing the frame size according to our need
    lip_crop = cv2.resize(lip_crop, (500, 300))
 
    # font which we will be using to display FPS
    font = cv2.FONT_HERSHEY_SIMPLEX
    # time when we finish processing for this frame
    new_frame_time = time.time()
    # Calculating the fps
    # fps will be number of frame processed in given time frame
    # since their will be most of time error of 0.001 second
    # we will be subtracting it to get more accurate result
    fps = 1/(new_frame_time-prev_frame_time)
    prev_frame_time = new_frame_time
    # converting the fps into integer
    fps = int(fps)
    print(fps)
    # converting the fps to string so that we can display it on frame
    # by using putText function
    fps = str(fps)
    # putting the FPS count on the frame
    cv2.putText(lip_crop, fps, (7, 70), font, 3, (100, 255, 0), 3, cv2.LINE_AA)
 
    # displaying the frame with fps
    cv2.imshow('frame', lip_crop) 
    cv2.waitKey(1)
#     if cv2.waitKey(50) == 13 :      # Specifying (Enter) button to break the loop
#         break
    count+=1
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()

#     print("count********** ", count)
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
#             gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            cv2.imshow('Gray image', img)
            images.append(img)
        print(img.shape)
    return images
```

```python

def convert_to_gray():
    path = glob.glob("data/*jpg")
    images = []
    for file in path:
        img = cv2.imread(file)
    #     cv2.imshow("Image", img)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        cv2.imshow('Gray image', gray)
        images.append(gray)
    cv2.waitKey(2000)
    return images

cv2.destroyAllWindows()
```
